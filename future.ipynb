```python
import csv
import pandas as pd

school_age_population = pd.read_csv('학령인구.csv')
new_ac = pd.read_csv('new_admissionCapacity.csv')

s_age = school_age_population

s_age = s_age[25:30].reset_index(drop=True)
s_age.drop(s_age.columns[0],axis=1,inplace=True)
s_age.set_index('인구종류별')
s_age.drop(1,axis=0,inplace=True)
s_age.drop(2,axis=0,inplace=True)
s_age.drop(3,axis=0,inplace=True)
s_age.reset_index(drop=True,inplace=True)

# 데이터프레임 행열 전환  .T
# s_age.transpose() 이거 안됨
s_age = s_age.T
# s_age.to_csv('학령인구_99~21.csv')
new_ac['합계']
# s_age.rename(index={'0':'학령인구(천명): 계(6-21세)','1':'학령인구(천명): 대학교(18-21세)'},inplace=True)
s_age.drop('인구종류별',axis=0,inplace=True)
s_age = s_age.astype(float)
s_age.reset_index(inplace=True)
s_age.columns=['연도','학령인구전체(6-21)','대학생학령인구(18-21)']
s_age.insert(loc=3,column='입학정원',value=new_ac['합계'])



```

```python
import matplotlib.pyplot as plt
import seaborn as sns

s_age
fig, ax = plt.subplots(1,2,figsize=(10,5))
s_age.plot.scatter(x='학령인구전체(6-21)',y='입학정원',ax=ax[0])
s_age.plot.scatter(x='대학생학령인구(18-21)',y='입학정원',ax=ax[1])




plt.show()
s_age
```
